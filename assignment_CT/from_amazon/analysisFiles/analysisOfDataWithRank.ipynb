## Question 3
### c. Relative Ranking

What: Relative ranking evaluates a brand's search visibility. Brands with SKUs appearing higher in search results are considered to have a better ranking.

Formula: Sum all SKU ranks for each brand, then divide by the number of SKUs to calculate the average rank for each brand.

Explanation: Lower average rank implies higher visibility, as products appear earlier in search results.

```python
# Calculate relative ranking for each brand
brand_ranking = df.groupby('Brand')['Rank'].mean().sort_values()

print("Relative Ranking of Brands:\n", brand_ranking)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
brand_ranking.plot(kind='bar', color='salmon')
plt.title('Relative Ranking of Brands')
plt.xlabel('Brand')
plt.ylabel('Average Rank')
plt.xticks(rotation=45)
plt.show()
```

## Question 4
### d. Relative Rating

What: Similar to ranking, relative rating evaluates a brand's product quality based on average customer ratings.

Formula: Sum of ratings for each brand divided by the number of SKUs.

```python
# Calculate relative rating for each brand
brand_rating = df.groupby('Brand')['Rating'].mean().sort_values(ascending=False)

print("Relative Rating of Brands:\n", brand_rating)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
brand_rating.plot(kind='bar', color='lightgreen')
plt.title('Relative Rating of Brands')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

## Question 5
### e. Price Distribution of SKUs

```python
# Define price ranges
bins = [0, 2999, 4999, 9999, 14999, 19999, float('inf')]
labels = ['<INR 3000', 'INR 3000-4999', 'INR 5000-9999', 'INR 10000-14999', 'INR 15000-19999', '>INR 20000']
price_distribution = pd.cut(df['Price'], bins=bins, labels=labels).value_counts().sort_index()

print("Price Distribution of SKUs:\n", price_distribution)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
price_distribution.plot(kind='bar', color='orange')
plt.title('Price Distribution of SKUs')
plt.xlabel('Price Range')
plt.ylabel('Number of SKUs')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file


# Set the aesthetic style of the plots
sns.set(style='whitegrid')

# Create a box plot to show price distribution by brand
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Brand', y='Price')

# Add title and labels
plt.title('Price Distribution by Brand', fontsize=16)
plt.xlabel('Brand', fontsize=14)
plt.ylabel('Price', fontsize=14)

# Rotate x labels for better readability if needed
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
bins = [0, 2999, 4999, 9999, 14999, 19999, float('inf')]
labels = ['<INR 3000', 'INR 3000-4999', 'INR 5000-9999', 'INR 10000-14999', 'INR 15000-19999', '>INR 20000']
df['Price_Category'] = pd.cut(df['Price'], bins=bins, labels=labels)

# Group by Brand and Price_Category and calculate average Rating and SKU count
brand_sales = df.groupby(['Price_Category', 'Brand']).agg(
    SKU_Count=('Product_Title', 'count'),    # SKU count for each brand in each price range
    Avg_Rating=('Rating', 'mean')            # Average rating for each brand in each price range
).reset_index()

# Find the brand with the highest SKU_Count in each price category
top_brand_by_price = brand_sales.loc[brand_sales.groupby('Price_Category')['SKU_Count'].idxmax()]

print("Top Brand by SKU Count in Each Price Category:")
print(top_brand_by_price)
```
