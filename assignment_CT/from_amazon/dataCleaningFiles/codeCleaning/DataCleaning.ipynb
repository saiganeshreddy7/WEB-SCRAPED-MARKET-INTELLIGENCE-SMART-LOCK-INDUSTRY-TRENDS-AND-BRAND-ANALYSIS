```python
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv('/Users/saiganeshreddykodekandla/Documents/projects/circuitTechnologies/circuitTechnologiesAmazon/csv_files/rawDataAmazonScrap.csv')
```

```python
df.head()
```

```python
# Fill all empty columns with 0
df.fillna(0, inplace=True)
```

```python
# Get the shape of the DataFrame (rows, columns)
shape = df.shape
shape
```

```python
# Remove duplicates while keeping one original
df = df.drop_duplicates(subset='Product Title', keep='first')
```

```python
df.head()
```

```python
# Get the shape of the DataFrame (rows, columns)
shape = df.shape
shape
```

```python
# Remove rupee symbol and commas from "Price" column and convert to float
df['Price'] = df['Price'].replace({'₹': '', 'INR': '', ',': ''}, regex=True).astype(float)
```

```python
df.head()
```

```python
# Keep only numeric ratings, replace others with 0
df['Rating'] = df['Rating'].apply(lambda x: x if isinstance(x, (int, float)) else (float(x) if str(x).replace('.', '', 1).isdigit() else 0))
```

```python
df.head()
```

```python
# Remove any non-numeric text from "Rating Count" and convert to integers
df['Rating Count'] = df['Rating Count'].str.replace(r'\D', '', regex=True)  # Keep only digits
df['Rating Count'] = df['Rating Count'].replace('', '0')  # Replace empty strings with '0'

# Convert to integer, replacing any NaN with 0 before conversion
df['Rating Count'] = df['Rating Count'].replace(pd.NA, '0').astype(int)  # Convert to integer
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/Users/saiganeshreddykodekandla/Documents/projects/circuitTechnologies/circuitTechnologiesAmazon/csv_files/afterstep8ratingcount.csv')

# Function to split the 'Best Seller Rank' data
def split_bestseller_rank(rank):
    if pd.isna(rank):
        return pd.NA, pd.NA
    parts = rank.split('#', 2)  # Split into at most 3 parts to isolate the second '#'
    if len(parts) < 3:
        return rank, pd.NA  # If there's no second '#', keep the original data
    return f'#{parts[1].strip()}', f'#{parts[2].strip()}'

# Apply the function to create two new columns
df['Best Seller Rank'], df['BestsellerRank'] = zip(*df['Best Seller Rank'].map(split_bestseller_rank))

# Save the modified DataFrame to a new CSV file
df.to_csv('/Users/saiganeshreddykodekandla/Documents/projects/circuitTechnologies/circuitTechnologiesAmazon/csv_files/modified_file.csv', index=False)
```

```python
import pandas as pd
import re

# Load the CSV file
df = pd.read_csv('/Users/saiganeshreddykodekandla/Documents/projects/circuitTechnologies/circuitTechnologiesAmazon/csv_files/modified_file.csv')

# Function to clean the 'Best Seller Rank' column
def clean_best_seller_rank(value):
    # Use regex to find all digits and join them without commas
    return ''.join(re.findall(r'\d+', str(value)))

# Apply the cleaning function to the column
df['Best Seller Rank'] = df['Best Seller Rank'].apply(clean_best_seller_rank)

# Save the cleaned DataFrame back to CSV
df.to_csv('/Users/saiganeshreddykodekandla/Documents/projects/circuitTechnologies/circuitTechnologiesAmazon/csv_files/cleaned_file.csv', index=False)
```
